# Retail Mini-Module: Sales/SF Sensitivity (Colab Template)

This notebook generates a sensitivity chart showing how Retail Value responds to changes in Sales per SF.

Required outputs:
- Exported notebook (.ipynb)
- PNG chart (retail_sales_sensitivity.png)
- Shareable Colab link (view access)


```python
import numpy as np
import matplotlib.pyplot as plt

# === ENTER YOUR ASSUMPTIONS (match Excel inputs) ===
RETAIL_SF = 20000
PERCENT_RENT = 0.08
VACANCY = 0.08
OPEX_RATIO = 0.30
CAP_RATE = 0.0675

assert RETAIL_SF > 0
assert 0 <= PERCENT_RENT <= 1
assert 0 <= VACANCY < 1
assert 0 <= OPEX_RATIO < 1
assert CAP_RATE > 0
```

```python
# Sensitivity range for Sales per SF
sales_psf = np.linspace(200, 800, 25)

# Retail value logic
gross_sales = sales_psf * RETAIL_SF
rent = gross_sales * PERCENT_RENT
egi = rent * (1 - VACANCY)
noi = egi * (1 - OPEX_RATIO)
value = noi / CAP_RATE
value_millions = value / 1_000_000

value_millions[:5]
```

```python
# Plot
plt.figure(figsize=(8,5))
plt.plot(sales_psf, value_millions, marker='o')
plt.xlabel('Sales per SF ($/SF/year)')
plt.ylabel('Retail Value (Millions $)')
plt.title('Retail Value Sensitivity to Sales Productivity')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Save PNG for GitHub
out_path = 'retail_sales_sensitivity.png'
plt.figure(figsize=(8,5))
plt.plot(sales_psf, value_millions, marker='o')
plt.xlabel('Sales per SF ($/SF/year)')
plt.ylabel('Retail Value (Millions $)')
plt.title('Retail Value Sensitivity to Sales Productivity')
plt.grid(True)
plt.tight_layout()
plt.savefig(out_path, dpi=200)
out_path
```
